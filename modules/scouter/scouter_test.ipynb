```python
from elasticsearch import Elasticsearch
es = Elasticsearch("155.230.34.145:9200")

query = {"size": 0, 
  "query": {
    "bool": {
      "must": [
        {"match": {
          "extContent": "삼성전자"
        }}
      ]
    }
  }, 
  "aggs": {
    "group_by_date": {
      "date_histogram": {
        "field": "postingDate",
        "calendar_interval": "day",
        "format": "yyyy-MM-dd"
      }
    }
  }
}

res = es.search(index='newspaper', body=query)
print(res)
```
